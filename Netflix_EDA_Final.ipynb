```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Load your dataset
count = pd.read_csv(r'C:\Users\pv437\Desktop\Data Scince Folder\Projects\Project 3\EDA\netflix_filled3.csv', encoding='latin1')
# Function to fetch country information from Wikipedia using BeautifulSoup
def fetch_country(title):
    url = f"https://en.wikipedia.org/wiki/{title.replace(' ', '_')}"
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        try:
            # Find the first paragraph containing country information
            country_info = soup.find('p', text=lambda text: 'country' in text.lower()).text
            country = country_info.split("country")[1].split('.')[0].strip()
            return country
        except:
            return None
    else:
        return None

# Update missing values in the "country" column
for index, row in count.iterrows():
    if pd.isnull(row['country']):
        title = row['title']
        country = fetch_country(title)
        count.at[index, 'country'] = country



```

```python
# Analyzing Cast and Directors for Recommendation
# Analyzing the most frequent cast members and directors can be useful for recommendation systems

# Function to count occurrences in a column with comma-separated values
def count_occurrences(dataframe, column):
    count_dict = {}
    for items in dataframe[column].dropna():
        for item in items.split(', '):
            if item in count_dict:
                count_dict[item] += 1
            else:
                count_dict[item] = 1
    return count_dict

# Counting occurrences for cast and directors
cast_count = count_occurrences(netflix, 'cast')
director_count = count_occurrences(netflix, 'director')

# Top 10 cast members and directors
top_10_cast = pd.DataFrame(list(cast_count.items()), columns=['Cast', 'Count']).sort_values(by='Count', ascending=False).head(10)
top_10_directors = pd.DataFrame(list(director_count.items()), columns=['Director', 'Count']).sort_values(by='Count', ascending=False).head(10)

# Plotting top 10 cast members and directors
fig, axes = plt.subplots(1, 2, figsize=(20, 8))
sns.barplot(ax=axes[0], x='Count', y='Cast', data=top_10_cast)
axes[0].set_title('Top 10 Cast Members')
sns.barplot(ax=axes[1], x='Count', y='Director', data=top_10_directors)
axes[1].set_title('Top 10 Directors')
plt.tight_layout()
plt.show()
```

```python
# Analyzing Genres for Recommendation
# Understanding popular genres can help in making better content recommendations

# Counting genre occurrences
genre_count = count_occurrences(netflix, 'listed_in')

# Top 10 genres
top_10_genres = pd.DataFrame(list(genre_count.items()), columns=['Genre', 'Count']).sort_values(by='Count', ascending=False).head(10)

# Plotting top 10 genres
plt.figure(figsize=(12, 6))
sns.barplot(x='Count', y='Genre', data=top_10_genres)
plt.title('Top 10 Genres on Netflix')
plt.show()
```

```python
# Collaborative Filtering Based Analysis
# Collaborative filtering is a method of making automatic predictions (filtering) about the interests of a user by collecting preferences from many users (collaborating).
# Analyzing the cast and director combinations can provide insights into successful collaborations.

# Creating a new column that combines cast and director
netflix['cast_director'] = netflix['cast'] + ' ' + netflix['director']

# Function to count unique combinations of cast and director
def count_combinations(dataframe, column):
    combination_dict = {}
    for items in dataframe[column].dropna():
        if items in combination_dict:
            combination_dict[items] += 1
        else:
            combination_dict[items] = 1
    return combination_dict

# Counting combinations
cast_director_combinations = count_combinations(netflix, 'cast_director')

# Top 10 cast-director combinations
top_10_combinations = pd.DataFrame(list(cast_director_combinations.items()), columns=['Combination', 'Count']).sort_values(by='Count', ascending=False).head(10)

# Plotting top 10 combinations
plt.figure(figsize=(12, 6))
sns.barplot(x='Count', y='Combination', data=top_10_combinations)
plt.title('Top 10 Cast-Director Combinations')
plt.show()
```

```python
from kmodes.kprototypes import KPrototypes
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import LabelEncoder
import numpy as np
import pandas as pd

# Assuming netflix_data is your dataset
# Convert categorical columns to string type to ensure compatibility
categorical_indices = [0, 1, 2, 3, 4, 6, 7, 8]
K_Value.iloc[:, categorical_indices] = K_Value.iloc[:, categorical_indices].astype(str)

# Initialize a LabelEncoder for each categorical column
label_encoders = [LabelEncoder() for _ in categorical_indices]

# Apply LabelEncoder to each categorical column
for i, col in enumerate(categorical_indices):
    K_Value.iloc[:, col] = label_encoders[i].fit_transform(K_Value.iloc[:, col])

# Convert DataFrame to dense numpy array
dense_data = K_Value.values

# Initialize a list to store silhouette scores for different numbers of clusters
silhouette_scores = []

# Range of cluster numbers to try
cluster_range = range(2, 11)

for n_clusters in cluster_range:
    # Initialize K-Prototypes with the current number of clusters
    kproto = KPrototypes(n_clusters=n_clusters, verbose=2, max_iter=20)
    
    # Fit the K-Prototypes model
    kproto.fit(dense_data, categorical=categorical_indices)
    
    # Compute silhouette score
    silhouette_avg = silhouette_score(dense_data, kproto.labels_)
    silhouette_scores.append(silhouette_avg)

# Find the optimal number of clusters based on the highest silhouette score
optimal_n_clusters = cluster_range[np.argmax(silhouette_scores)]

print("Optimal number of clusters:", optimal_n_clusters)
```

```python
import matplotlib.pyplot as plt

# Plot silhouette scores
plt.plot(cluster_range, silhouette_scores, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs. Number of Clusters')
plt.show()
```

```python
import numpy as np
from kmodes.kprototypes import KPrototypes
from sklearn.metrics import silhouette_score
import pandas as pd

def compute_gap_statistic(data, categorical_indices, max_clusters=10, n_refs=5, random_state=None):
    
    if isinstance(data, pd.DataFrame):
        data = data.values

    # Initialize arrays to store within-cluster dispersions
    within_dispersion = np.zeros(max_clusters)
    within_dispersion_ref = np.zeros((max_clusters, n_refs))

    # Loop over different number of clusters
    for k in range(1, max_clusters + 1):
        # Initialize K-Prototypes model
        kproto = KPrototypes(n_clusters=k, random_state=random_state)

        # Fit the model
        kproto.fit(data, categorical=categorical_indices)

        # Compute within-cluster dispersion
        within_dispersion[k - 1] = kproto.cost_

        # Generate reference datasets and compute within-cluster dispersion
        for i in range(n_refs):
            # Generate random reference dataset
            ref_data = np.random.rand(*data.shape)

            # Fit K-Prototypes model to reference dataset
            kproto_ref = KPrototypes(n_clusters=k, random_state=random_state)
            kproto_ref.fit(ref_data, categorical=categorical_indices)

            # Compute within-cluster dispersion for reference dataset
            within_dispersion_ref[k - 1, i] = kproto_ref.cost_

    # Compute Gap Statistics
    gap_stats = np.log(within_dispersion_ref.mean(axis=1)) - np.log(within_dispersion)

    return gap_stats

# Example usage:
# Assuming netflix_data is your dataset
# Convert categorical columns to string type to ensure compatibility
categorical_indices = [0, 1, 2, 3, 4, 6, 7, 8]
K_Value.iloc[:, categorical_indices] = K_Value.iloc[:, categorical_indices].astype(str)

# Compute Gap Statistics
gap_stats = compute_gap_statistic(K_Value, categorical_indices, max_clusters=10, n_refs=5, random_state=42)

```

```python
# Plot Gap Statistics
plt.plot(range(1, len(gap_stats) + 1), gap_stats, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Gap Statistic')
plt.title('Gap Statistic vs. Number of Clusters')
plt.show()
```
